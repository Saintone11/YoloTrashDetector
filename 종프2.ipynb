```python
from re import X
import json
import glob
import os

for filepath in glob.glob('/content/drive/MyDrive/trashproject/*.json'):
  with open(os.path.join(os.getcwd(), filepath), 'r') as f:
      data = json.load(f)
  print(filepath)
  screenwidth = data['asset']['parent']['size']['width']
  screenheight = data['asset']['parent']['size']['height']
  imagename = data['asset']['name']
  regions = data['regions']
  file = open(imagename+".txt", "w")
  for region in regions:
    tags = str(region['tags'][0])
    if 'PET' in tags:
      tagnum = 0
    elif 'PS' in tags:
      tagnum = 1
    elif 'PP' in tags:
      tagnum = 2
    elif 'Others' in tags:
      tagnum = 3
    elif 'CAN' in tags:
      tagnum = 4
    elif 'GLASS' in tags:
      tagnum = 5
    elif 'Undefined' in  tags:
      tagnum = 6
    x_min = region['boundingBox']['left']
    y_min = region['boundingBox']['top']
    width = region['boundingBox']['width']
    height = region['boundingBox']['height']
    x_min = x_min / screenwidth
    y_min = y_min / screenheight
    width = width /screenwidth
    height = height / screenheight
    print(tagnum, x_min, y_min, width, height)
    file.write(f"{tagnum:d} {x_min:f} {y_min:f} {width:f} {height:f}\n")
  file.close()
```
